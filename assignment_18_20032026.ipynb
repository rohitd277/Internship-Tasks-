*Description* :Collect 20 messy sentences and identify slang, emojis, typos; explain preprocessing needed.

```python
import pandas as pd

sentences = [
    "Hey bro 😎 what's up!!!",
    "I luv this movie soooo much!!!",
    "LOL that was funny 😂😂",
    "Gud mrng frnds",
    "This is amazng!!!",
    "Pls send me the details ASAP!!!",
    "I'm sooo happpy today 😊",
    "Thx for ur help",
    "OMG!!! unbelievable 😲",
    "C u soon!!!",
    "Wanna go out??",
    "This product is gr8",
    "Hahaha 😂😂😂",
    "U r amazing!!!",
    "Lets meet tmrw",
    "Cant wait!!! 😍",
    "Sry for the delay",
    "Best day everrrr",
    "Yessss!!! finally done",
    "Noooooo way!!!"
]

df = pd.DataFrame({"Text": sentences})

# Identify issues
def analyze(text):
    slang = any(word in text.lower() for word in ["luv","bro","u","ur","tmrw","gr8"])
    emoji = any(char in text for char in ["😂","😎","😊","😍","😲"])
    typo = any(word in text.lower() for word in ["gud","happpy","amazng","sry"])
    
    return pd.Series([slang, emoji, typo])

df[["Slang","Emoji","Typo"]] = df["Text"].apply(analyze)

print(df)
```
