```python
num = int(input("Enter a number: ")) #print the number is odd or even 

if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

```python
total = 0 #sum until the user stops giving input 

num = int(input("Enter number: "))

while num != 0:
    total = total + num
    num = int(input("Enter number: "))

print("Total Sum =", total)
```
