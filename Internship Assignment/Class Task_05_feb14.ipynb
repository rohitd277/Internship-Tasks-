```python
import pandas as pd

# Step 1: Read CSV file
df = pd.read_csv("students.csv")

# Step 2: Handle missing values
df = df.fillna(0)

# Step 3: Add Total column
df["Total"] = df["Maths"] + df["Science"] + df["English"]

# Step 4: Add Average column
df["Average"] = df["Total"] / 3

# Step 5: Top 3 students based on total
top_students = df.sort_values(by="Total", ascending=False).head(3)

print("Top 3 Students:")
print(top_students)

# Step 6: Department wise average marks
dept_avg = df.groupby("Department")[["Maths","Science","English"]].mean()

print("\nDepartment Wise Average Marks:")
print(dept_avg)

# Step 7: Students scoring above 75 in all subjects
high_scorers = df[(df["Maths"]>75) & (df["Science"]>75) & (df["English"]>75)]

print("\nStudents scoring above 75 in all subjects:")
print(high_scorers)
```
