```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Step 1: Create dataset with missing values and duplicates
data = {
    "Name": ["Rohit", "Anita", "Rahul", "Priya", "Rohit"],
    "Age": [25, 30, None, 28, 25],
    "Salary": [50000, None, 45000, 52000, 50000],
    "City": ["Hyderabad", "hyderabad", "HYDERABAD", "Mumbai", "Hyderabad"],
    "Experience": [2, 5, 3, None, 2]
}

df = pd.DataFrame(data)

print("Original Dataset:\n")
print(df)

# Step 2: Remove duplicate rows
df = df.drop_duplicates()

# Step 3: Handle missing values using column mean
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())
df["Experience"] = df["Experience"].fillna(df["Experience"].mean())

# Step 4: Standardize city names
df["City"] = df["City"].str.lower().str.capitalize()

# Step 5: Apply Min-Max scaling to Age and Salary
scaler = MinMaxScaler()

df[["Age", "Salary"]] = scaler.fit_transform(df[["Age", "Salary"]])

# Step 6: Display final cleaned dataset
print("\nFinal Cleaned Dataset:\n")
print(df)
```
