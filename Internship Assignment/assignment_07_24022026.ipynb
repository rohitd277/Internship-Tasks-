```python
# Import libraries
import pandas as pd

# Load dataset
data = pd.read_csv("student_data.csv")

# Display top rows
print("Top 5 Rows of Dataset:")
print(data.head())

# Find highest value in each column
print("\nHighest Values in Each Column:")
print(data.max(numeric_only=True))

# Count missing values
print("\nMissing Values in Dataset:")
print(data.isnull().sum())

# Insights
print("\nInsights from Dataset:")

print("1. Student with highest marks:")
print(data.loc[data['Marks'].idxmax()])

print("\n2. Average study hours:", data['StudyHours'].mean())

print("\n3. Average marks:", data['Marks'].mean())

print("\n4. Student with highest attendance:")
print(data.loc[data['Attendance'].idxmax()])

print("\n5. Correlation between Study Hours and Marks:")
print(data[['StudyHours','Marks']].corr())
```
