```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Step 1: Create dataset
data = {
    "Engine_Size": [1000,1200,1500,1800,2000],
    "Mileage": [18,20,15,12,10],
    "Age": [5,4,3,2,1],
    "Price": [500000,600000,750000,900000,1100000]
}

df = pd.DataFrame(data)

print("Dataset:\n")
print(df)

# Step 2: Define features and target
X = df[["Engine_Size","Mileage","Age"]]
y = df["Price"]

# Step 3: Train model
model = LinearRegression()
model.fit(X,y)

# Step 4: Create new data for prediction (with feature names)
new_car = pd.DataFrame({
    "Engine_Size":[1500],
    "Mileage":[20],
    "Age":[3]
})

# Step 5: Predict price
predicted_price = model.predict(new_car)

print("\nPredicted Price:", predicted_price)

# Step 6: Print coefficients
coef_table = pd.DataFrame({
    "Feature": X.columns,
    "Coefficient": model.coef_
})

print("\nModel Coefficients:")
print(coef_table)
```
