*Description* : Explain Netflix-like recommendations using KNN and create a small similarity example.

```python
import numpy as np
from sklearn.neighbors import NearestNeighbors

# Take number of movies
n = int(input("Enter number of movies: "))

movies = []
movie_names = []

# Take movie data from user
for i in range(n):
    print("\nEnter details for movie", i+1)
    
    name = input("Movie name: ")
    action = int(input("Action rating (1-5): "))
    comedy = int(input("Comedy rating (1-5): "))
    romance = int(input("Romance rating (1-5): "))
    
    movie_names.append(name)
    movies.append([action, comedy, romance])

movies = np.array(movies)

# Train KNN model
model = NearestNeighbors(n_neighbors=2, metric='euclidean')
model.fit(movies)

print("\nMovie Index List:")
for i in range(len(movie_names)):
    print(i, "-", movie_names[i])

# Select movie for recommendation
movie_index = int(input("\nEnter movie index to find similar movie: "))

distances, indices = model.kneighbors([movies[movie_index]])

print("\nRecommended movie similar to", movie_names[movie_index], ":")

for i in indices[0]:
    if i != movie_index:
        print(movie_names[i])
```
