*Description* : Build a Python program that takes name, age, hobby and prints a personalized message while categorizing age using conditionals.

```python
name = input("Enter your name: ")
age = int(input("Enter your age: "))
hobby = input("Enter your hobby: ")

if age < 13:
    category = "a child"
elif age < 20:
    category = "a teenager"
elif age < 60:
    category = "an adult"
else:
    category = "a senior citizen"

print("\nHello", name + "!")
print("You are", age, "years old and you are", category + ".")
print("It's nice to know that your hobby is", hobby + "!")
```
