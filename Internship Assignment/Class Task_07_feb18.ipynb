```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Create dataset
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Revenue": [20000, 25000, 27000, 30000, 32000, 35000],
    "Marketing_Spent": [5000, 6000, 6500, 7000, 7200, 7500],
    "Profit": [8000, 10000, 11000, 13000, 14000, 16000]
}

df = pd.DataFrame(data)

print("Dataset:")
print(df)

# Step 2: Line Plot for Revenue
plt.figure()
plt.plot(df["Month"], df["Revenue"], marker='o')
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()

# Step 3: Scatter Plot (Marketing Spent vs Profit)
plt.figure()
plt.scatter(df["Marketing_Spent"], df["Profit"])
plt.title("Marketing Spent vs Profit")
plt.xlabel("Marketing Spent")
plt.ylabel("Profit")
plt.show()

# Step 4: Correlation Heatmap
corr_matrix = df[["Revenue", "Marketing_Spent", "Profit"]].corr()

plt.figure()
sns.heatmap(corr_matrix, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
