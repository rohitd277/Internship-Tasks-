*Description* : Draw a decision tree predicting whether you should play outside.

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import pandas as pd
import matplotlib.pyplot as plt

# Dataset: Weather conditions and decision to play outside
data = {
    "Weather": ["Sunny","Sunny","Overcast","Rain","Rain","Rain","Overcast","Sunny"],
    "Temperature": ["Hot","Hot","Hot","Mild","Cool","Cool","Mild","Cool"],
    "Humidity": ["High","High","High","High","Normal","Normal","Normal","High"],
    "Wind": ["Weak","Strong","Weak","Weak","Weak","Strong","Strong","Weak"],
    "Play": ["No","No","Yes","Yes","Yes","No","Yes","No"]
}

df = pd.DataFrame(data)

# Convert categorical data to numbers
df_encoded = pd.get_dummies(df.drop("Play", axis=1))
y = df["Play"]

# Train Decision Tree
model = DecisionTreeClassifier()
model.fit(df_encoded, y)

# Plot the Decision Tree
plt.figure(figsize=(12,6))
plot_tree(model, feature_names=df_encoded.columns, class_names=model.classes_, filled=True)
plt.title("Decision Tree: Should You Play Outside?")
plt.show()
```
