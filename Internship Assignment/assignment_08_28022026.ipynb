```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Student': ['Amit','Riya','John','Sara','David','Priya','Kiran','Anu','Rahul','Meena'],
    'StudyHours': [2,5,1,4,6,3,7,2,5,4],
    'Marks': [50,78,40,72,88,65,92,55,80,75]
}

df = pd.DataFrame(data)

print(df)
```

```python
plt.bar(df['Student'], df['Marks'])
plt.title("Student Marks Comparison")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
plt.pie(df['StudyHours'], labels=df['Student'], autopct='%1.1f%%')
plt.title("Study Hours Distribution")
plt.show()
```

```python
plt.hist(df['Marks'], bins=5)
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.show()
```

```python
print("DATA STORY")

print("1. Students who studied more hours generally scored higher marks.")
print("2. Kiran studied the most hours and achieved the highest marks.")
print("3. Students who studied less than 3 hours scored comparatively lower marks.")
print("4. The histogram shows most students scored between 70 and 90 marks.")
print("5. Overall trend shows a positive relationship between study hours and marks.")
```
