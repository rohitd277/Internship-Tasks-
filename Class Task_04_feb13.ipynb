```python
import numpy as np

# Step 1: Create a list of temperatures
temperature = [28, 32, 30, 37, 26, 38]

# Step 2: Convert list into NumPy array
temp_array = np.array(temperature)

# Step 3: Print maximum temperature
print("Maximum Temperature:", np.max(temp_array))

# Step 4: Print minimum temperature
print("Minimum Temperature:", np.min(temp_array))

# Step 5: Calculate average temperature
print("Average Temperature:", np.mean(temp_array))

# Step 6: Display last three days temperature
print("Last three days temperature:", temp_array[-3:])
```
