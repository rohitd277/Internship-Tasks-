```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Dataset (15+ sentences)
sentences = [
    "i like playing football",
    "football is a great sport",
    "i love cricket",
    "cricket is popular in india",
    "players practice daily",
    "sports keep us healthy",
    "i enjoy watching matches",
    "team won the match",
    "he scored a goal",
    "she likes tennis",
    "tennis is fun",
    "they play basketball",
    "basketball is fast",
    "training improves performance",
    "coach guides players"
]

# Convert text to vectors (TF-IDF)
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(sentences)

# Get feature names (words)
words = vectorizer.get_feature_names_out()

# Convert matrix to array
X_array = X.toarray()

# Choose a word
word = "football"

# Find index of word
word_index = list(words).index(word)

# Get vector of that word (column)
word_vector = X_array[:, word_index]

print("Vector for word 'football':\n", word_vector)

# Find similarity between words
similarities = cosine_similarity(X_array.T)

# Get similar words
word_idx = list(words).index(word)
sim_scores = list(enumerate(similarities[word_idx]))

# Sort by similarity
sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

print("\nTop 5 similar words to 'football':")
for i in sim_scores[1:6]:
    print(words[i[0]], ":", i[1])
```
